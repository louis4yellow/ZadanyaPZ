```python
import random
import string

def generate_password(length, use_letters=True, use_digits=True, use_special=True, count=1):

    chars = ""

    if use_letters:
        chars += string.ascii_letters
    if use_digits:
        chars += string.digits
    if use_special:
        chars += string.punctuation

    if not chars:
        return ["–û—à–∏–±–∫–∞: –≤—ã–±–µ—Ä–∏—Ç–µ —Ö–æ—Ç—è –±—ã –æ–¥–∏–Ω —Ç–∏–ø —Å–∏–º–≤–æ–ª–æ–≤!"]

    passwords = []
    for _ in range(count):
        password = ''.join(random.choice(chars) for _ in range(length))
        passwords.append(password)

    return passwords

def check_password_strength(password):
    """
    –ü—Ä–æ–≤–µ—Ä—è–µ—Ç —Å–ª–æ–∂–Ω–æ—Å—Ç—å –ø–∞—Ä–æ–ª—è –∏ –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç –æ—Ü–µ–Ω–∫—É –∏ —Ä–µ–∫–æ–º–µ–Ω–¥–∞—Ü–∏–∏

    –í–æ–∑–≤—Ä–∞—â–∞–µ—Ç: (–æ—Ü–µ–Ω–∫–∞, —Å–ø–∏—Å–æ–∫ —Ä–µ–∫–æ–º–µ–Ω–¥–∞—Ü–∏–π)
    –æ—Ü–µ–Ω–∫–∞: "–û—á–µ–Ω—å —Å–ª–∞–±—ã–π", "–°–ª–∞–±—ã–π", "–°—Ä–µ–¥–Ω–∏–π", "–ù–∞–¥–µ–∂–Ω—ã–π", "–û—á–µ–Ω—å –Ω–∞–¥–µ–∂–Ω—ã–π"
    """
    score = 0
    recommendations = []

    length = len(password)
    if length < 8:
        recommendations.append("–£–≤–µ–ª–∏—á—å—Ç–µ –¥–ª–∏–Ω—É –ø–∞—Ä–æ–ª—è –¥–æ 8+ —Å–∏–º–≤–æ–ª–æ–≤")
    elif length < 10:
        score += 1
    elif length < 12:
        score += 2
    else:
        score += 3

    if any(c.isupper() for c in password):
        score += 1
    else:
        recommendations.append("–î–æ–±–∞–≤—å—Ç–µ –±—É–∫–≤—ã –≤–µ—Ä—Ö–Ω–µ–≥–æ —Ä–µ–≥–∏—Å—Ç—Ä–∞")

    if any(c.islower() for c in password):
        score += 1
    else:
        recommendations.append("–î–æ–±–∞–≤—å—Ç–µ –±—É–∫–≤—ã –Ω–∏–∂–Ω–µ–≥–æ —Ä–µ–≥–∏—Å—Ç—Ä–∞")

    if any(c.isdigit() for c in password):
        score += 1
    else:
        recommendations.append("–î–æ–±–∞–≤—å—Ç–µ —Ü–∏—Ñ—Ä—ã")

    if any(c in string.punctuation for c in password):
        score += 1
    else:
        recommendations.append("–î–æ–±–∞–≤—å—Ç–µ —Å–ø–µ—Ü–∏–∞–ª—å–Ω—ã–µ —Å–∏–º–≤–æ–ª—ã")

    if score <= 2:
        strength = "–û—á–µ–Ω—å —Å–ª–∞–±—ã–π"
    elif score <= 3:
        strength = "–°–ª–∞–±—ã–π"
    elif score <= 4:
        strength = "–°—Ä–µ–¥–Ω–∏–π"
    elif score <= 5:
        strength = "–ù–∞–¥–µ–∂–Ω—ã–π"
    else:
        strength = "–û—á–µ–Ω—å –Ω–∞–¥–µ–∂–Ω—ã–π"

    return strength, recommendations

def main():
    """–û—Å–Ω–æ–≤–Ω–∞—è —Ñ—É–Ω–∫—Ü–∏—è –ø—Ä–æ–≥—Ä–∞–º–º—ã"""
    print("=" * 50)
    print(" –ì–ï–ù–ï–†–ê–¢–û–† –ë–ï–ó–û–ü–ê–°–ù–´–• –ü–ê–†–û–õ–ï–ô")
    print("=" * 50)

    while True:
        print("\n--- –ù–∞—Å—Ç—Ä–æ–π–∫–∏ –≥–µ–Ω–µ—Ä–∞—Ü–∏–∏ ---")

        try:
            length = int(input("–î–ª–∏–Ω–∞ –ø–∞—Ä–æ–ª—è (–º–∏–Ω–∏–º—É–º 6): "))
            if length < 6:
                print("–î–ª–∏–Ω–∞ –ø–∞—Ä–æ–ª—è –¥–æ–ª–∂–Ω–∞ –±—ã—Ç—å –Ω–µ –º–µ–Ω–µ–µ 6 —Å–∏–º–≤–æ–ª–æ–≤. –£—Å—Ç–∞–Ω–æ–≤–ª–µ–Ω–æ –∑–Ω–∞—á–µ–Ω–∏–µ 6.")
                length = 6
        except ValueError:
            print("–û—à–∏–±–∫–∞: –≤–≤–µ–¥–∏—Ç–µ —á–∏—Å–ª–æ. –£—Å—Ç–∞–Ω–æ–≤–ª–µ–Ω–∞ –¥–ª–∏–Ω–∞ 8.")
            length = 8

        print("\n–í—ã–±–µ—Ä–∏—Ç–µ —Ç–∏–ø—ã —Å–∏–º–≤–æ–ª–æ–≤ –¥–ª—è –ø–∞—Ä–æ–ª—è:")
        use_letters = input("–ò—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –±—É–∫–≤—ã? (–¥–∞/–Ω–µ—Ç): ").lower() in ['–¥–∞', '–¥', 'yes', 'y', '']
        use_digits = input("–ò—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å —Ü–∏—Ñ—Ä—ã? (–¥–∞/–Ω–µ—Ç): ").lower() in ['–¥–∞', '–¥', 'yes', 'y', '']
        use_special = input("–ò—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å —Å–ø–µ—Ü–∏–∞–ª—å–Ω—ã–µ —Å–∏–º–≤–æ–ª—ã? (–¥–∞/–Ω–µ—Ç): ").lower() in ['–¥–∞', '–¥', 'yes', 'y', '']
        if not (use_letters or use_digits or use_special):
            print("–û—à–∏–±–∫–∞: –≤—ã–±–µ—Ä–∏—Ç–µ —Ö–æ—Ç—è –±—ã –æ–¥–∏–Ω —Ç–∏–ø —Å–∏–º–≤–æ–ª–æ–≤!")
            continue

        try:
            count = int(input("\n–°–∫–æ–ª—å–∫–æ –ø–∞—Ä–æ–ª–µ–π —Å–≥–µ–Ω–µ—Ä–∏—Ä–æ–≤–∞—Ç—å? (–ø–æ —É–º–æ–ª—á–∞–Ω–∏—é 1): ") or "1")
            if count < 1:
                count = 1
        except ValueError:
            print("–û—à–∏–±–∫–∞: –≤–≤–µ–¥–∏—Ç–µ —á–∏—Å–ª–æ. –°–≥–µ–Ω–µ—Ä–∏—Ä–æ–≤–∞–Ω 1 –ø–∞—Ä–æ–ª—å.")
            count = 1

        print(f"\n{'='*50}")
        print(f"–ì–ï–ù–ï–†–ê–¶–ò–Ø {count} –ü–ê–†–û–õ–ï–ô (–¥–ª–∏–Ω–∞: {length})")
        print(f"{'='*50}")

        passwords = generate_password(length, use_letters, use_digits, use_special, count)

        for i, password in enumerate(passwords, 1):
            print(f"\n–ü–∞—Ä–æ–ª—å {i}: {password}")

            strength, recommendations = check_password_strength(password)
            print(f"–°–ª–æ–∂–Ω–æ—Å—Ç—å: {strength}")

            if recommendations:
                print("–†–µ–∫–æ–º–µ–Ω–¥–∞—Ü–∏–∏:")
                for rec in recommendations[:3]:
                    print(f"  ‚Ä¢ {rec}")

        again = input("\n–°–≥–µ–Ω–µ—Ä–∏—Ä–æ–≤–∞—Ç—å –µ—â–µ –ø–∞—Ä–æ–ª–∏? (–¥–∞/–Ω–µ—Ç): ").lower()
        if again not in ['–¥–∞', '–¥', 'yes', 'y', '']:
            print("\n–°–ø–∞—Å–∏–±–æ –∑–∞ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ –≥–µ–Ω–µ—Ä–∞—Ç–æ—Ä–∞ –ø–∞—Ä–æ–ª–µ–π!")
            break

if __name__ == "__main__":
    main()
```

```python
from datetime import datetime
from collections import defaultdict
from typing import List, Dict, Optional

class CafeOrderSystem:
    def __init__(self):
        self.menu = {}
        self.orders = []
        self.next_menu_id = 1
        self.next_order_id = 1
        self.dish_popularity = defaultdict(int)

    class Dish:
        def __init__(self, dish_id, name, price, category, prep_time):
            self.id = dish_id
            self.name = name
            self.price = price
            self.category = category
            self.prep_time = prep_time

        def __repr__(self):
            return f"{self.name} ({self.category}) - {self.price}‚ÇΩ, {self.prep_time}–º–∏–Ω"

    class Order:
        def __init__(self, order_id, table_number, dishes):
            self.id = order_id
            self.table_number = table_number
            self.dishes = dishes
            self.status = "–ü—Ä–∏–Ω—è—Ç"
            self.created_at = datetime.now()
            self.completed_at = None

        def calculate_total(self):
            return sum(dish.price for dish in self.dishes)

        def complete_order(self):
            self.status = "–û–ø–ª–∞—á–µ–Ω"
            self.completed_at = datetime.now()

        def __repr__(self):
            return f"–ó–∞–∫–∞–∑ #{self.id} (–°—Ç–æ–ª {self.table_number}): {self.status}, —Å—É–º–º–∞: {self.calculate_total()}‚ÇΩ"

    def add_dish(self, name: str, price: float, category: str, prep_time: int) -> int:
        dish_id = self.next_menu_id
        dish = self.Dish(dish_id, name, price, category, prep_time)
        self.menu[dish_id] = dish
        self.next_menu_id += 1
        print(f"‚úì –ë–ª—é–¥–æ '{name}' –¥–æ–±–∞–≤–ª–µ–Ω–æ –≤ –º–µ–Ω—é (ID: {dish_id})")
        return dish_id

    def create_order(self, table_number: int, dish_ids: List[int]) -> Optional[int]:
        dishes = []
        for dish_id in dish_ids:
            if dish_id not in self.menu:
                print(f"‚úó –ë–ª—é–¥–æ —Å ID {dish_id} –Ω–µ –Ω–∞–π–¥–µ–Ω–æ –≤ –º–µ–Ω—é")
                return None
            dishes.append(self.menu[dish_id])

        order_id = self.next_order_id
        order = self.Order(order_id, table_number, dishes)
        self.orders.append(order)
        self.next_order_id += 1

        for dish in dishes:
            self.dish_popularity[dish.id] += 1

        print(f"‚úì –ó–∞–∫–∞–∑ #{order_id} —Å–æ–∑–¥–∞–Ω –¥–ª—è —Å—Ç–æ–ª–∞ {table_number}")
        return order_id

    def update_order_status(self, order_id: int, status: str) -> bool:
        for order in self.orders:
            if order.id == order_id:
                order.status = status
                print(f"‚úì –°—Ç–∞—Ç—É—Å –∑–∞–∫–∞–∑–∞ #{order_id} –∏–∑–º–µ–Ω–µ–Ω –Ω–∞ '{status}'")
                return True
        print(f"‚úó –ó–∞–∫–∞–∑ #{order_id} –Ω–µ –Ω–∞–π–¥–µ–Ω")
        return False

    def calculate_order_total(self, order_id: int) -> Optional[float]:
        """–í–æ–∑–≤—Ä–∞—â–∞–µ—Ç —Å—Ç–æ–∏–º–æ—Å—Ç—å –∑–∞–∫–∞–∑–∞ –ø–æ –µ–≥–æ ID"""
        for order in self.orders:
            if order.id == order_id:
                total = order.calculate_total()
                print(f"–°—Ç–æ–∏–º–æ—Å—Ç—å –∑–∞–∫–∞–∑–∞ #{order_id}: {total}‚ÇΩ")
                return total
        print(f"‚úó –ó–∞–∫–∞–∑ #{order_id} –Ω–µ –Ω–∞–π–¥–µ–Ω")
        return None

    def get_popular_dishes(self, top_n: int = 5) -> List[tuple]:
        """
        –í–æ–∑–≤—Ä–∞—â–∞–µ—Ç —Ç–æ–ø N —Å–∞–º—ã—Ö –ø–æ–ø—É–ª—è—Ä–Ω—ã—Ö –±–ª—é–¥
        :param top_n: –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –±–ª—é–¥ –≤ —Ç–æ–ø–µ
        :return: —Å–ø–∏—Å–æ–∫ –∫–æ—Ä—Ç–µ–∂–µ–π (–Ω–∞–∑–≤–∞–Ω–∏–µ, –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∑–∞–∫–∞–∑–æ–≤, –∫–∞—Ç–µ–≥–æ—Ä–∏—è)
        """
        if not self.dish_popularity:
            print("–ù–µ—Ç –¥–∞–Ω–Ω—ã—Ö –æ –ø–æ–ø—É–ª—è—Ä–Ω–æ—Å—Ç–∏ –±–ª—é–¥")
            return []

        sorted_dishes = sorted(
            self.dish_popularity.items(),
            key=lambda x: x[1],
            reverse=True
        )[:top_n]

        result = []
        print(f"\nüî• –¢–û–ü-{top_n} —Å–∞–º—ã—Ö –ø–æ–ø—É–ª—è—Ä–Ω—ã—Ö –±–ª—é–¥:")
        for i, (dish_id, count) in enumerate(sorted_dishes, 1):
            dish = self.menu[dish_id]
            result.append((dish.name, count, dish.category))
            print(f"{i}. {dish.name} ({dish.category}) - –∑–∞–∫–∞–∑–∞–Ω–æ {count} —Ä–∞–∑")

        return result

    def get_revenue_by_category(self) -> Dict[str, float]:
        revenue = defaultdict(float)

        for order in self.orders:
            if order.status == "–û–ø–ª–∞—á–µ–Ω":
                for dish in order.dishes:
                    revenue[dish.category] += dish.price

        print("\nüí∞ –í—ã—Ä—É—á–∫–∞ –ø–æ –∫–∞—Ç–µ–≥–æ—Ä–∏—è–º:")
        for category, amount in revenue.items():
            print(f"{category}: {amount:.2f}‚ÇΩ")

        return dict(revenue)

    def show_menu(self):
        """–ü–æ–∫–∞–∑—ã–≤–∞–µ—Ç –≤—Å–µ –±–ª—é–¥–∞ –≤ –º–µ–Ω—é"""
        print("\nüìã –ú–µ–Ω—é:")
        if not self.menu:
            print("–ú–µ–Ω—é –ø—É—Å—Ç–æ")
            return

        by_category = defaultdict(list)
        for dish in self.menu.values():
            by_category[dish.category].append(dish)

        for category, dishes in by_category.items():
            print(f"\n{category}:")
            for dish in dishes:
                print(f"  [{dish.id}] {dish.name} - {dish.price}‚ÇΩ ({dish.prep_time} –º–∏–Ω)")

    def show_orders(self, show_all=True):
        """–ü–æ–∫–∞–∑—ã–≤–∞–µ—Ç –≤—Å–µ –∑–∞–∫–∞–∑—ã"""
        print("\nüì¶ –ó–∞–∫–∞–∑—ã:")
        if not self.orders:
            print("–ù–µ—Ç –∑–∞–∫–∞–∑–æ–≤")
            return

        for order in self.orders:
            if show_all or order.status != "–û–ø–ª–∞—á–µ–Ω":
                print(f"\n{order}")
                for dish in order.dishes:
                    print(f"  - {dish.name}")

def demo():
    cafe = CafeOrderSystem()

    print("=" * 50)
    print("–î–û–ë–ê–í–õ–ï–ù–ò–ï –ë–õ–Æ–î –í –ú–ï–ù–Æ")
    print("=" * 50)

    cafe.add_dish("–≠—Å–ø—Ä–µ—Å—Å–æ", 150, "–ù–∞–ø–∏—Ç–∫–∏", 5)
    cafe.add_dish("–ö–∞–ø—É—á–∏–Ω–æ", 250, "–ù–∞–ø–∏—Ç–∫–∏", 7)
    cafe.add_dish("–ë–æ—Ä—â", 350, "–°—É–ø—ã", 15)
    cafe.add_dish("–°–æ–ª—è–Ω–∫–∞", 400, "–°—É–ø—ã", 20)
    cafe.add_dish("–¶–µ–∑–∞—Ä—å —Å –∫—É—Ä–∏—Ü–µ–π", 450, "–°–∞–ª–∞—Ç—ã", 10)
    cafe.add_dish("–ì—Ä–µ—á–µ—Å–∫–∏–π —Å–∞–ª–∞—Ç", 380, "–°–∞–ª–∞—Ç—ã", 8)
    cafe.add_dish("–°—Ç–µ–π–∫ –†–∏–±–∞–π", 1200, "–ì–æ—Ä—è—á–µ–µ", 25)
    cafe.add_dish("–ü–∞—Å—Ç–∞ –ö–∞—Ä–±–æ–Ω–∞—Ä–∞", 550, "–ì–æ—Ä—è—á–µ–µ", 15)
    cafe.add_dish("–¢–∏—Ä–∞–º–∏—Å—É", 350, "–î–µ—Å–µ—Ä—Ç—ã", 10)
    cafe.add_dish("–ß–∏–∑–∫–µ–π–∫", 320, "–î–µ—Å–µ—Ä—Ç—ã", 5)

    cafe.show_menu()

    print("\n" + "=" * 50)
    print("–°–û–ó–î–ê–ù–ò–ï –ó–ê–ö–ê–ó–û–í")
    print("=" * 50)

    order1 = cafe.create_order(5, [1, 2, 9])
    cafe.update_order_status(order1, "–ì–æ—Ç–æ–≤–∏—Ç—Å—è")
    cafe.update_order_status(order1, "–ì–æ—Ç–æ–≤")
    cafe.update_order_status(order1, "–û–ø–ª–∞—á–µ–Ω")

    order2 = cafe.create_order(3, [3, 4, 7])
    cafe.update_order_status(order2, "–û–ø–ª–∞—á–µ–Ω")

    order3 = cafe.create_order(2, [5, 6, 8, 2])
    cafe.update_order_status(order3, "–û–ø–ª–∞—á–µ–Ω")

    order4 = cafe.create_order(1, [1, 1, 9, 10])
    cafe.update_order_status(order4, "–û–ø–ª–∞—á–µ–Ω")

    cafe.show_orders()

    print("\n" + "=" * 50)
    print("–†–ê–°–ß–ï–¢ –°–¢–û–ò–ú–û–°–¢–ò")
    print("=" * 50)
    cafe.calculate_order_total(order1)
    cafe.calculate_order_total(order2)

    print("\n" + "=" * 50)
    print("–ü–û–ü–£–õ–Ø–†–ù–´–ï –ë–õ–Æ–î–ê")
    print("=" * 50)
    cafe.get_popular_dishes(5)

    print("\n" + "=" * 50)
    print("–°–¢–ê–¢–ò–°–¢–ò–ö–ê –í–´–†–£–ß–ö–ò")
    print("=" * 50)
    revenue = cafe.get_revenue_by_category()

    total_revenue = sum(revenue.values())
    print(f"\nüí∞ –û–ë–©–ê–Ø –í–´–†–£–ß–ö–ê: {total_revenue:.2f}‚ÇΩ")

if __name__ == "__main__":
    demo()
```

```python
from collections import defaultdict
from datetime import datetime

class CafeSystem:
    def __init__(self):
        self.menu, self.orders, self.popular = {}, [], defaultdict(int)
        self.mid = self.oid = 1

    def add_dish(self, name, price, cat, time):
        self.menu[self.mid] = {'id':self.mid, 'name':name, 'price':price, 'cat':cat, 'time':time}
        print(f"–î–æ–±–∞–≤–ª–µ–Ω–æ: {name}")
        self.mid += 1
        return self.mid-1

    def create_order(self, table, dish_ids):
        dishes = []
        for did in dish_ids:
            if did not in self.menu: print(f"ID {did} –Ω–µ –Ω–∞–π–¥–µ–Ω"); return None
            dishes.append(self.menu[did])
            self.popular[did] += 1

        order = {'id':self.oid, 'table':table, 'dishes':dishes, 'status':'–ü—Ä–∏–Ω—è—Ç', 'time':datetime.now()}
        self.orders.append(order)
        self.oid += 1
        print(f"–ó–∞–∫–∞–∑ #{order['id']} —Å–æ–∑–¥–∞–Ω")
        return order['id']

    def update_status(self, oid, status):
        for o in self.orders:
            if o['id'] == oid: o['status'] = status; print(f"–°—Ç–∞—Ç—É—Å #{oid}: {status}"); return True
        print("–ù–µ –Ω–∞–π–¥–µ–Ω"); return False

    def order_total(self, oid):
        for o in self.orders:
            if o['id'] == oid: total = sum(d['price'] for d in o['dishes']); print(f"–°—É–º–º–∞: {total} —Ä—É–±."); return total
        print("–ù–µ –Ω–∞–π–¥–µ–Ω"); return None

    def popular_dishes(self, n=5):
        top = sorted(self.popular.items(), key=lambda x: x[1], reverse=True)[:n]
        print(f"\n–¢–æ–ø-{n}:")
        for i, (did, cnt) in enumerate(top, 1): print(f"{i}. {self.menu[did]['name']} - {cnt}")
        return [(self.menu[did]['name'], cnt) for did, cnt in top]

    def revenue_by_cat(self):
        rev = defaultdict(float)
        for o in self.orders:
            if o['status'] == '–û–ø–ª–∞—á–µ–Ω':
                for d in o['dishes']: rev[d['cat']] += d['price']
        print("\n–í—ã—Ä—É—á–∫–∞:")
        for cat, amt in rev.items(): print(f"{cat}: {amt} —Ä—É–±.")
        return dict(rev)

    def show_menu(self):
        print("\n–ú–µ–Ω—é:")
        if not self.menu: print("–ü—É—Å—Ç–æ"); return
        cat = defaultdict(list)
        for d in self.menu.values(): cat[d['cat']].append(d)
        for c, items in cat.items():
            print(f"\n{c}:")
            for i in items: print(f"  [{i['id']}] {i['name']} - {i['price']} —Ä—É–±.")

    def show_orders(self):
        print("\n–ó–∞–∫–∞–∑—ã:")
        if not self.orders: print("–ù–µ—Ç"); return
        for o in self.orders:
            total = sum(d['price'] for d in o['dishes'])
            print(f"\n#{o['id']} (–°—Ç–æ–ª {o['table']}): {o['status']}, {total} —Ä—É–±.")
            for d in o['dishes']: print(f"  - {d['name']}")

def demo():
    c = CafeSystem()

    print("="*30)
    print("–î–û–ë–ê–í–õ–ï–ù–ò–ï")
    print("="*30)
    ids = []
    for item in [("–≠—Å–ø—Ä–µ—Å—Å–æ",150,"–ù–∞–ø–∏—Ç–∫–∏",5), ("–ö–∞–ø—É—á–∏–Ω–æ",250,"–ù–∞–ø–∏—Ç–∫–∏",7), ("–ë–æ—Ä—â",350,"–°—É–ø—ã",15),
                 ("–°–æ–ª—è–Ω–∫–∞",400,"–°—É–ø—ã",20), ("–¶–µ–∑–∞—Ä—å",450,"–°–∞–ª–∞—Ç—ã",10), ("–ì—Ä–µ—á–µ—Å–∫–∏–π",380,"–°–∞–ª–∞—Ç—ã",8),
                 ("–°—Ç–µ–π–∫",1200,"–ì–æ—Ä—è—á–µ–µ",25), ("–ü–∞—Å—Ç–∞",550,"–ì–æ—Ä—è—á–µ–µ",15), ("–¢–∏—Ä–∞–º–∏—Å—É",350,"–î–µ—Å–µ—Ä—Ç—ã",10),
                 ("–ß–∏–∑–∫–µ–π–∫",320,"–î–µ—Å–µ—Ä—Ç—ã",5)]:
        ids.append(c.add_dish(*item))

    c.show_menu()

    print("\n"+"="*30)
    print("–ó–ê–ö–ê–ó–´")
    print("="*30)
    o1 = c.create_order(5, [1,2,9])
    c.update_status(o1, "–û–ø–ª–∞—á–µ–Ω")
    o2 = c.create_order(3, [3,4,7])
    c.update_status(o2, "–û–ø–ª–∞—á–µ–Ω")
    o3 = c.create_order(2, [5,6,8,2])
    c.update_status(o3, "–û–ø–ª–∞—á–µ–Ω")
    o4 = c.create_order(1, [1,1,9,10])
    c.update_status(o4, "–û–ø–ª–∞—á–µ–Ω")

    c.show_orders()

    print("\n"+"="*30)
    print("–°–¢–ê–¢–ò–°–¢–ò–ö–ê")
    print("="*30)
    c.order_total(o1)
    c.popular_dishes(3)
    rev = c.revenue_by_cat()
    print(f"\n–ò–¢–û–ì–û: {sum(rev.values())} —Ä—É–±.")

if __name__ == "__main__":
    demo()
```

```python
import re

def validate_inn():
    while True:
        inn = input("–í–≤–µ–¥–∏—Ç–µ –ò–ù–ù (12 —Ü–∏—Ñ—Ä): ").strip()
        if re.fullmatch(r'\d{12}', inn):
            print("–ò–ù–ù –∫–æ—Ä—Ä–µ–∫—Ç–µ–Ω.")
            return inn
        else:
            print("–û—à–∏–±–∫–∞: –ò–ù–ù –¥–æ–ª–∂–µ–Ω —Å–æ—Å—Ç–æ—è—Ç—å —Ä–æ–≤–Ω–æ –∏–∑ 12 —Ü–∏—Ñ—Ä. –ü–æ–ø—Ä–æ–±—É–π—Ç–µ —Å–Ω–æ–≤–∞.")

def validate_price():
    while True:
        price_input = input("–í–≤–µ–¥–∏—Ç–µ —Ü–µ–Ω—É —Ç–æ–≤–∞—Ä–∞ (–ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–æ–µ —á–∏—Å–ª–æ –¥–æ 1 000 000): ").strip()
        try:
            price = float(price_input)
            if price <= 0:
                print("–û—à–∏–±–∫–∞: —Ü–µ–Ω–∞ –¥–æ–ª–∂–Ω–∞ –±—ã—Ç—å –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–æ–π.")
            elif price > 1_000_000:
                print("–û—à–∏–±–∫–∞: —Ü–µ–Ω–∞ –Ω–µ –¥–æ–ª–∂–Ω–∞ –ø—Ä–µ–≤—ã—à–∞—Ç—å 1 000 000.")
            else:
                print("–¶–µ–Ω–∞ –∫–æ—Ä—Ä–µ–∫—Ç–Ω–∞.")
                return price
        except ValueError:
            print("–û—à–∏–±–∫–∞: –≤–≤–µ–¥–∏—Ç–µ –∫–æ—Ä—Ä–µ–∫—Ç–Ω–æ–µ —á–∏—Å–ª–æ (—Ü–µ–ª–æ–µ –∏–ª–∏ –¥–µ—Å—è—Ç–∏—á–Ω–æ–µ).")

def validate_password():
    while True:
        password = input("–í–≤–µ–¥–∏—Ç–µ –ø–∞—Ä–æ–ª—å (8-20 —Å–∏–º–≤–æ–ª–æ–≤, —Ö–æ—Ç—è –±—ã –æ–¥–Ω–∞ –∑–∞–≥–ª–∞–≤–Ω–∞—è –±—É–∫–≤–∞ –∏ –æ–¥–Ω–∞ —Ü–∏—Ñ—Ä–∞): ")
        if len(password) < 8 or len(password) > 20:
            print("–û—à–∏–±–∫–∞: –¥–ª–∏–Ω–∞ –ø–∞—Ä–æ–ª—è –¥–æ–ª–∂–Ω–∞ –±—ã—Ç—å –æ—Ç 8 –¥–æ 20 —Å–∏–º–≤–æ–ª–æ–≤.")
        elif not re.search(r'[A-Z]', password):
            print("–û—à–∏–±–∫–∞: –ø–∞—Ä–æ–ª—å –¥–æ–ª–∂–µ–Ω —Å–æ–¥–µ—Ä–∂–∞—Ç—å —Ö–æ—Ç—è –±—ã –æ–¥–Ω—É –∑–∞–≥–ª–∞–≤–Ω—É—é –±—É–∫–≤—É.")
        elif not re.search(r'\d', password):
            print("–û—à–∏–±–∫–∞: –ø–∞—Ä–æ–ª—å –¥–æ–ª–∂–µ–Ω —Å–æ–¥–µ—Ä–∂–∞—Ç—å —Ö–æ—Ç—è –±—ã –æ–¥–Ω—É —Ü–∏—Ñ—Ä—É.")
        else:
            print("–ü–∞—Ä–æ–ª—å –∫–æ—Ä—Ä–µ–∫—Ç–µ–Ω.")
            return password

print("–ü—Ä–æ–≤–µ—Ä–∫–∞ –ò–ù–ù:")
validate_inn()
print("\n–ü—Ä–æ–≤–µ—Ä–∫–∞ —Ü–µ–Ω—ã:")
validate_price()
print("\n–ü—Ä–æ–≤–µ—Ä–∫–∞ –ø–∞—Ä–æ–ª—è:")
validate_password()
```
